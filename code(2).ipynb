```python
def cap_outliers_iqr(df, col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    df[col] = df[col].clip(lower=lower_bound, upper=upper_bound)
    return df
```

```python
def classify_bmi(bmi):
    if bmi < 18.5:
        return 'Underweight'
    elif bmi < 25:
        return 'Normal'
    elif bmi < 30:
        return 'Overweight'
    else:
        return 'Obese'  # includes all obesity levels

df_numerical['bmi_category'] = df_numerical['BMI'].apply(classify_bmi)
```

```python
# Unrealistic systolic
df_numerical[(df_numerical['ap_hi'] < 80) | (df_numerical['ap_hi'] > 250)]

# Unrealistic diastolic
df_numerical[(df_numerical['ap_lo'] < 40) | (df_numerical['ap_lo'] > 150)]

# Extremely high pulse pressure
df_numerical[df_numerical['pulse_pressure'] > 100]
```

```python

def classify_bp(row):
    ap_hi = row['ap_hi']
    ap_lo = row['ap_lo']

    if ap_hi >= 180 or ap_lo >= 120:
        return 'Hypertensive Crisis'
    elif ap_hi >= 140 or ap_lo >= 90:
        return 'Stage 2 Hypertension'
    elif ap_hi >= 130 or ap_lo >= 80:
        return 'Stage 1 Hypertension'
    elif ap_hi >= 120 and ap_lo < 80:
        return 'Elevated'
    elif ap_hi < 120 and ap_lo < 80:
        return 'Normal'
    else:
        return 'Other/Unclassified'
```

```python
df_numerical['bp_category'] = df_numerical.apply(classify_bp, axis=1)
```

```python
def classify_pulse_pressure(pp):
    if pp > 60:
        return 'Wide Pulse Pressure'
    elif pp < 30:
        return 'Narrow Pulse Pressure'
    else: # pp >= 30 and pp <= 60
        return 'Normal Pulse Pressure'
```

```python
df_numerical['pulse_pressure_category'] = df_numerical['pulse_pressure'].apply(classify_pulse_pressure)
```

```python
ordinal_features_with_order = {
    'bmi_category': ['Underweight', 'Normal', 'Overweight', 'Obese'], # Sorted by increasing risk
    'age_group' : ['young_adult','middle_aged','senior'],
    'bp_category': ['Normal', 'Elevated', 'Stage 1 Hypertension', 'Stage 2 Hypertension', 'Hypertensive Crisis'], # Clinically defined order
    'pulse_pressure_category': ['Normal Pulse Pressure', 'Narrow Pulse Pressure', 'Wide Pulse Pressure'], # Order based on deviation/risk
    'gluc': [1, 2, 3],
    'cholesterol': [1, 2, 3]
}
```

```python
ordinal_cols = list(ordinal_features_with_order.keys())
```

```python
nominal_cols = [
    'gender',
    'smoke',
    'alco',
    'active'
]
```

```python
# Define column transformers
preprocessor = ColumnTransformer(transformers=[
    ('num', StandardScaler(), numerical_cols),
    ('ord', OrdinalEncoder(categories=[ordinal_features_with_order[col] for col in ordinal_cols],
                           handle_unknown='use_encoded_value', unknown_value=-1), ordinal_cols),
    ('nom', OneHotEncoder(handle_unknown='ignore', sparse_output=False), nominal_cols)
])
```

```python
# Fit only on training data
preprocessor.fit(X_train)

# Transform datasets
X_train_final = preprocessor.transform(X_train)
X_test_final = preprocessor.transform(X_test)
```

```python
def get_feature_names(column_transformer):
    output_features = []
    for name, preprocessor, features in column_transformer.transformers_:
        if hasattr(preprocessor, 'get_feature_names_out'):
            if isinstance(features, str):
                features = [features]
            if name == 'ord':
                output_features.extend([f"{f}_ordinal" for f in features])
            elif name == 'nom':
                output_features.extend(preprocessor.get_feature_names_out(features))
            elif name == 'num':
                output_features.extend([f"{f}_scaled" for f in features])
        else:
            if isinstance(features, str):
                features = [features]
            output_features.extend([f"{f}_{name}" for f in features])
    return output_features

feature_names_final = get_feature_names(preprocessor)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score
from sklearn.model_selection import train_test_split

# Assuming you have: X_train_final, y_train, X_test_final, y_test

# Adjusted hyperparameters to reduce overfitting
rf_regularized = RandomForestClassifier(
    n_estimators=100,           # Number of trees
    max_depth=10,               # Limit depth of each tree
    min_samples_split=10,       # More samples required to split an internal node
    min_samples_leaf=5,         # Minimum samples required at a leaf node
    max_features='sqrt',        # Use sqrt of total features at each split
    random_state=42,
    n_jobs=-1
)

# Fit the model
rf_regularized.fit(X_train_final, y_train)

# Predict on train and test
y_train_pred = rf_regularized.predict(X_train_final)
y_test_pred = rf_regularized.predict(X_test_final)

# Evaluate performance
train_acc = accuracy_score(y_train, y_train_pred)
test_acc = accuracy_score(y_test, y_test_pred)

print("=== Regularized Random Forest Performance ===")
print(f"Train Accuracy: {train_acc:.4f}")
print(f"Test Accuracy: {test_acc:.4f}")
print("\nTest Classification Report:")
print(classification_report(y_test, y_test_pred))
```

```python
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
scoring = ['accuracy', 'roc_auc', 'f1_weighted']

print("\n=== Cross-Validation Results (Before Tuning) ===")
for name, model in models.items():
    print(f"\n--- {name} CV ---")
    scores = cross_validate(model, X_train_final, y_train, cv=cv, scoring=scoring, n_jobs=-1)
    for metric in scoring:
        print(f"{metric.upper()} - Mean: {scores[f'test_{metric}'].mean():.4f}, Std: {scores[f'test_{metric}'].std():.4f}")
```

```python
def tune_model(objective_func, model_name, n_trials=50):
    print(f"\n--- Tuning {model_name} ---")
    study = optuna.create_study(direction='maximize', sampler=optuna.samplers.TPESampler(seed=42))
    study.optimize(objective_func, n_trials=n_trials, show_progress_bar=True)
    print(f"Best AUC-ROC: {study.best_value:.4f}")
    print("Best hyperparameters:")
    for k, v in study.best_params.items():
        print(f"   {k}: {v}")
    return study.best_params

def objective_rf(trial):
    params = {
        'n_estimators': trial.suggest_int('n_estimators', 50, 500),
        'max_depth': trial.suggest_int('max_depth', 5, 30),
        'min_samples_split': trial.suggest_int('min_samples_split', 2, 20),
        'min_samples_leaf': trial.suggest_int('min_samples_leaf', 1, 10),
        'max_features': trial.suggest_categorical('max_features', ['sqrt', 'log2', 0.8, 1.0]),
        'random_state': 42,
        'n_jobs': -1
    }
    model = RandomForestClassifier(**params)
    aucs = []
    X_np = X_train_final.values if hasattr(X_train_final, 'values') else X_train_final
    y_np = y_train.values if hasattr(y_train, 'values') else y_train
    for train_idx, val_idx in cv.split(X_np, y_np):
        model.fit(X_np[train_idx], y_np[train_idx])
        y_val_proba = model.predict_proba(X_np[val_idx])[:, 1]
        aucs.append(roc_auc_score(y_np[val_idx], y_val_proba))
    return np.mean(aucs)

def objective_xgb(trial):
    params = {
        'n_estimators': trial.suggest_int('n_estimators', 50, 1000),
        'max_depth': trial.suggest_int('max_depth', 3, 15),
        'learning_rate': trial.suggest_float('learning_rate', 0.01, 0.3, log=True),
        'subsample': trial.suggest_float('subsample', 0.6, 1.0),
        'colsample_bytree': trial.suggest_float('colsample_bytree', 0.6, 1.0),
        'gamma': trial.suggest_float('gamma', 0.0, 0.5),
        'reg_alpha': trial.suggest_float('reg_alpha', 1e-5, 1.0, log=True),
        'reg_lambda': trial.suggest_float('reg_lambda', 1e-5, 1.0, log=True),
        'use_label_encoder': False,
        'eval_metric': 'logloss',
        'random_state': 42,
        'n_jobs': -1
    }
    model = XGBClassifier(**params)
    aucs = []
    X_np = X_train_final.values if hasattr(X_train_final, 'values') else X_train_final
    y_np = y_train.values if hasattr(y_train, 'values') else y_train
    for train_idx, val_idx in cv.split(X_np, y_np):
        model.fit(X_np[train_idx], y_np[train_idx])
        y_val_proba = model.predict_proba(X_np[val_idx])[:, 1]
        aucs.append(roc_auc_score(y_np[val_idx], y_val_proba))
    return np.mean(aucs)

def objective_cat(trial):
    params = {
        'iterations': trial.suggest_int('iterations', 50, 1000),
        'learning_rate': trial.suggest_float('learning_rate', 0.01, 0.3, log=True),
        'depth': trial.suggest_int('depth', 3, 10),
        'l2_leaf_reg': trial.suggest_float('l2_leaf_reg', 1e-5, 10.0, log=True),
        'subsample': trial.suggest_float('subsample', 0.6, 1.0),
        'verbose': 0,
        'random_state': 42
    }
    model = CatBoostClassifier(**params)
    aucs = []
    X_np = X_train_final.values if hasattr(X_train_final, 'values') else X_train_final
    y_np = y_train.values if hasattr(y_train, 'values') else y_train
    for train_idx, val_idx in cv.split(X_np, y_np):
        model.fit(X_np[train_idx], y_np[train_idx])
        y_val_proba = model.predict_proba(X_np[val_idx])[:, 1]
        aucs.append(roc_auc_score(y_np[val_idx], y_val_proba))
    return np.mean(aucs)

params_rf = tune_model(objective_rf, "Random Forest")
params_xgb = tune_model(objective_xgb, "XGBoost")
params_cat = tune_model(objective_cat, "CatBoost")
```

```python
tuned_models = {
    'Random Forest': RandomForestClassifier(**params_rf, random_state=42, n_jobs=-1),
    'XGBoost': XGBClassifier(**params_xgb, use_label_encoder=False, eval_metric='logloss', random_state=42, n_jobs=-1),
    'CatBoost': CatBoostClassifier(**params_cat, verbose=0, random_state=42)
}

results_after_tuning = {}

for name, model in tuned_models.items():
    print(f"\n--- {name} Tuned ---")
    model.fit(X_train_final, y_train)
    y_pred = model.predict(X_test_final)
    y_proba = model.predict_proba(X_test_final)[:, 1]

    accuracy = accuracy_score(y_test, y_pred)
    roc_auc = roc_auc_score(y_test, y_proba)
    conf_mat = confusion_matrix(y_test, y_pred)
    class_report = classification_report(y_test, y_pred)

    print(f"Accuracy: {accuracy:.4f}")
    print(f"AUC-ROC: {roc_auc:.4f}")
    print("Confusion Matrix:\n", conf_mat)
    print("Classification Report:\n", class_report)

    results_after_tuning[name] = {
        'accuracy': accuracy,
        'roc_auc': roc_auc,
        'y_pred': y_pred,
        'y_proba': y_proba,
        'confusion_matrix': conf_mat
    }

    plt.figure(figsize=(6, 5))
    sns.heatmap(conf_mat, annot=True, fmt='d', cmap='Blues', cbar=False,
                xticklabels=['Predicted 0', 'Predicted 1'],
                yticklabels=['Actual 0', 'Actual 1'])
    plt.title(f'Confusion Matrix for {name} (Tuned)')
    plt.xlabel('Predicted Label')
    plt.ylabel('True Label')
    plt.show()
```

```python
print("\n=== Starting Stacking Ensemble ===")

rf_tuned = RandomForestClassifier(
    n_estimators=params_rf['n_estimators'], max_depth=params_rf['max_depth'],
    min_samples_split=params_rf['min_samples_split'], min_samples_leaf=params_rf['min_samples_leaf'],
    max_features=params_rf['max_features'], random_state=42, n_jobs=-1)

xgb_tuned = XGBClassifier(
    n_estimators=params_xgb['n_estimators'], max_depth=params_xgb['max_depth'],
    learning_rate=params_xgb['learning_rate'], subsample=params_xgb['subsample'],
    colsample_bytree=params_xgb['colsample_bytree'], gamma=params_xgb['gamma'],
    reg_alpha=params_xgb['reg_alpha'], reg_lambda=params_xgb['reg_lambda'],
    use_label_encoder=False, eval_metric='logloss', random_state=42, n_jobs=-1)

cat_tuned = CatBoostClassifier(
    iterations=params_cat['iterations'], learning_rate=params_cat['learning_rate'],
    depth=params_cat['depth'], l2_leaf_reg=params_cat['l2_leaf_reg'],
    subsample=params_cat['subsample'], verbose=0, random_state=42)

base_models = {
    'Random Forest': rf_tuned,
    'XGBoost': xgb_tuned,
    'CatBoost': cat_tuned
}

meta_model = LogisticRegression(solver='liblinear', random_state=42, n_jobs=-1, class_weight='balanced')
n_splits_oof = 5
skf_oof = StratifiedKFold(n_splits=n_splits_oof, shuffle=True, random_state=42)

oof_preds = np.zeros((X_train_final.shape[0], len(base_models)))
test_preds_base = np.zeros((X_test_final.shape[0], len(base_models)))

print(f"\nGenerating Out-Of-Fold predictions using {n_splits_oof}-Fold Stratified Cross-Validation...")

X_train_final_np = X_train_final.values if hasattr(X_train_final, 'values') else X_train_final
y_train_np = y_train.values if hasattr(y_train, 'values') else y_train

for i, (name, model) in enumerate(base_models.items()):
    print(f"   Processing base model: {name}")
    temp_test_preds = np.zeros((X_test_final.shape[0], n_splits_oof))

    for fold, (train_idx, val_idx) in enumerate(skf_oof.split(X_train_final_np, y_train_np)):
        X_train_fold, y_train_fold = X_train_final_np[train_idx], y_train_np[train_idx]
        X_val_fold = X_train_final_np[val_idx]

        model.fit(X_train_fold, y_train_fold)
        oof_preds[val_idx, i] = model.predict_proba(X_val_fold)[:, 1]
        temp_test_preds[:, fold] = model.predict_proba(X_test_final)[:, 1]

    test_preds_base[:, i] = temp_test_preds.mean(axis=1)

print("\nOOF and base test probabilities generated.")

X_meta_train = oof_preds
X_meta_test = test_preds_base

print(f"Shape of X_meta_train: {X_meta_train.shape}")
print(f"Shape of X_meta_test: {X_meta_test.shape}")

print("\nTraining the Meta-Model (Logistic Regression)...")
meta_model.fit(X_meta_train, y_train)

y_stacked_proba = meta_model.predict_proba(X_meta_test)[:, 1]
y_stacked_pred = (y_stacked_proba > 0.5).astype(int)

print("\n=== Stacking Ensemble Performance on Test Set ===")
accuracy_stacked = accuracy_score(y_test, y_stacked_pred)
roc_auc_stacked = roc_auc_score(y_test, y_stacked_proba)
conf_mat_stacked = confusion_matrix(y_test, y_stacked_pred)
class_report_stacked = classification_report(y_test, y_stacked_pred)

print(f"Test Accuracy: {accuracy_stacked:.4f}")
print("Confusion Matrix:\n", conf_mat_stacked)
print("Classification Report:\n", class_report_stacked)
print(f"AUC-ROC Score: {roc_auc_stacked:.4f}")

results_after_tuning['Stacking Ensemble'] = {
    'accuracy': accuracy_stacked,
    'roc_auc': roc_auc_stacked,
    'y_pred': y_stacked_pred,
    'y_proba': y_stacked_proba,
    'confusion_matrix': conf_mat_stacked
}

plt.figure(figsize=(6, 5))
sns.heatmap(conf_mat_stacked, annot=True, fmt='d', cmap='Blues', cbar=False,
            xticklabels=['Predicted 0', 'Predicted 1'],
            yticklabels=['Actual 0', 'Actual 1'])
plt.title('Confusion Matrix for Stacking Ensemble')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()

print("\nStacking ensemble process complete.")
```
